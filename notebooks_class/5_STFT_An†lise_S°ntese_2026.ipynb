### Exemplo de áudio

```python
x, sr = librosa.load(librosa.ex('trumpet'))
```

```python
ipd.Audio(x, rate = sr)
```

```python
plt.figure(figsize=(10, 4))
plt.title('Signal')
plt.tight_layout()
librosa.display.waveshow(x, sr=sr, color='r')

plt.tight_layout()
plt.show()
```

```python
D = librosa.stft(x, n_fft = 1024,  win_length = 1024, hop_length=512,  window='hann')
x_istft_phase = librosa.istft(D, hop_length=512,  window='hann')
```

```python
plt.figure(figsize=(10, 4))
plt.title('Magnitude-Phase istft reconstruction')
plt.tight_layout()
librosa.display.waveshow(x_istft_phase, sr = sr, color='r')

plt.tight_layout()
plt.show()
```

```python
# Get the magnitude of stft
S = np.abs(D)

# Invert without estimating phase
x_istft = librosa.istft(S, hop_length=512,  window='hann')

# Invert using Griffin-Lim
x_inv = librosa.griffinlim(S, hop_length=512,  window='hann')
```

```python
fig, ax = plt.subplots(nrows=4, sharex=True, sharey=True,figsize=(10,12))

librosa.display.waveshow(x, sr=sr, color='b', ax=ax[0])
ax[0].set(title='Original', xlabel=None)
ax[0].label_outer()

librosa.display.waveshow(x_inv, sr=sr, color='g', ax=ax[1] )
ax[1].set(title='Griffin-Lim reconstruction', xlabel=None)
ax[1].label_outer()

librosa.display.waveshow(x_istft, sr=sr, color='r', ax=ax[2])
ax[2].set_title('Magnitude-only istft reconstruction')
ax[2].label_outer()

librosa.display.waveshow(x_istft_phase, sr=sr, color='r', ax=ax[3])
ax[3].set_title('Magnitude-Phase istft reconstruction')

plt.tight_layout()
plt.show()
```

```python
# Griffin-Lim reconstruction
ipd.Audio(x_inv, rate = sr)
```

```python
# Magnitude-only istft reconstruction
ipd.Audio(x_istft, rate = sr)
```

```python
# Magnitude-Phase istft reconstruction
ipd.Audio(x_istft_phase, rate = sr)
```

## 2. Harmonic-Percussive Source Separation

```python
x, sr = librosa.load('wav\\Weeknd_Blinding_Lights.mp3', sr = None, offset = 14, duration = 13)
ipd.Audio(x, rate = sr)
```

```python
hop_length = 512
n_fft = 1024
n_window = 1024

X = librosa.stft(x, win_length = n_window, n_fft = n_fft, hop_length = hop_length, window = 'hann')
```

```python
# Perform harmonic-percussive source separation:
X_harmonic, X_percussive = librosa.decompose.hpss(X)

x_harmonic = librosa.istft(X_harmonic, win_length = n_window, hop_length = hop_length, window = 'hann')

x_percussive = librosa.istft(X_percussive, win_length = n_window, hop_length = hop_length, window = 'hann')
```

```python
ipd.Audio(x_percussive, rate = sr)
```

```python
ipd.Audio(x_harmonic, rate = sr)
```
